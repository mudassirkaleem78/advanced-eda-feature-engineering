```python
# ==========================================
# Create Backup Copy
# ==========================================

data = df.copy()

print("Backup Created Successfully")
```

```python
# ==========================================
# Missing Values
# ==========================================

missing_values = pd.DataFrame({

    "Missing Values": data.isnull().sum(),

    "Percentage": round(
        (data.isnull().sum()/len(data))*100,
        2
    )

})

missing_values.sort_values(
    by="Percentage",
    ascending=False
)
```

```python
numerical_columns = data.select_dtypes(
    include=["int64","float64"]
).columns
```

```python
for column in numerical_columns:

    data[column].fillna(
        data[column].median(),
        inplace=True
    )
```

```python
categorical_columns = data.select_dtypes(
    include="object"
).columns
```

```python
for column in categorical_columns:

    data[column].fillna(
        data[column].mode()[0],
        inplace=True
    )
```

```python
data.isnull().sum()
```

```python
data["Date"] = pd.to_datetime(
    data["Date"]
)
```

```python
data["Year"] = data["Date"].dt.year
```

```python
data["Month"] = data["Date"].dt.month
```

```python
data["Day"] = data["Date"].dt.day
```

```python
data["DayOfWeek"] = data["Date"].dt.day_name()
```

```python
data["IsWeekend"] = data["DayOfWeek"].isin(
    ["Saturday","Sunday"]
).astype(int)
```

```python
data["OrderValue"] = (
    data["Quantity"] *
    data["UnitPrice"]
)
```

```python
data["CouponUsed"] = (
    data["CouponCode"] != "No Coupon"
).astype(int)
```

```python
data["AverageItemPrice"] = (

    data["OrderValue"] /

    data["Quantity"]

)
```

```python
drop_columns = [

    "OrderID",

    "CustomerID",

    "TrackingNumber",

    "ShippingAddress",

    "Date"

]
```

```python
data.drop(
    columns=drop_columns,
    inplace=True
)
```

```python
numerical_columns = data.select_dtypes(
    include=["int64","float64"]
).columns
```

```python
for column in numerical_columns:

    Q1 = data[column].quantile(0.25)

    Q3 = data[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR

    upper = Q3 + 1.5 * IQR

    data[column] = np.where(

        data[column] < lower,

        lower,

        np.where(

            data[column] > upper,

            upper,

            data[column]

        )

    )
```

```python
categorical_columns = data.select_dtypes(
    include="object"
).columns
```

```python
encoder = LabelEncoder()

for column in categorical_columns:

    data[column] = encoder.fit_transform(
        data[column]
    )
```

```python
X = data.drop(
    "OrderStatus",
    axis=1
)

y = data["OrderStatus"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(

    X,

    y,

    test_size=0.20,

    random_state=42,

    stratify=y

)
```

```python
# ==========================================
# Additional Libraries
# ==========================================

from sklearn.pipeline import Pipeline
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
# ==========================================
# Machine Learning Models
# ==========================================

models = {

    "Logistic Regression": LogisticRegression(
        random_state=42,
        max_iter=1000
    ),

    "Decision Tree": DecisionTreeClassifier(
        random_state=42
    ),

    "Random Forest": RandomForestClassifier(
        n_estimators=300,
        random_state=42
    ),

    "Gradient Boosting": GradientBoostingClassifier(
        n_estimators=200,
        learning_rate=0.05,
        random_state=42
    )

}
```

```python
# ==========================================
# Evaluation Function
# ==========================================

def evaluate_model(model, X_train, X_test, y_train, y_test):

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)

    precision = precision_score(
        y_test,
        y_pred,
        average="weighted"
    )

    recall = recall_score(
        y_test,
        y_pred,
        average="weighted"
    )

    f1 = f1_score(
        y_test,
        y_pred,
        average="weighted"
    )

    return accuracy, precision, recall, f1, y_pred
```

```python
# ==========================================
# Train All Models
# ==========================================

results = []

predictions = {}

trained_models = {}
```

```python
for model_name, model in models.items():
    accuracy, precision, recall, f1, y_pred = evaluate_model(
        model,
        X_train,
        X_test,
        y_train,
        y_test
    )
    trained_models[model_name] = model
    predictions[model_name] = y_pred
    results.append({
        "Model": model_name,
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1
    })
    print("="*60)
    print(model_name)
    print("="*60)
    print(classification_report(y_test, y_pred))
```
